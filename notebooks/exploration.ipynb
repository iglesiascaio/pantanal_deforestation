```python
from runner import io 
import pandas as pd
```

```python
df_land_use = io.load_table("raw", "land_use")
df_land_use_transitions = io.load_table("raw", "land_use_transitions")
```

```python
list_municipios_pantanal = [
        "Barão de Melgaço (MT)",
        "Cáceres (MT)",
        "Itiquira (MT)",
        "Lambari D'Oeste (MT)",
        "Nossa Senhora do Livramento (MT)",
        "Poconé (MT)",
        "Santo Antônio do Leverger (MT)",
        "Aquidauana (MS)",
        "Bodoquena (MS)",
        "Corumbá (MS)",
        "Coxim (MS)",
        "Ladário (MS)",
        "Miranda (MS)",
        "Sonora (MS)",
        "Porto Murtinho (MS)",
        "Rio Verde de Mato Grosso (MS)",
    ]

df_land_use_transitions["location"] = df_land_use_transitions["city"] + " (" + df_land_use_transitions["state"] + ")"

df_land_use_transitions_pantanal = df_land_use_transitions.query("location.isin(@list_municipios_pantanal)")
```

```python
assert len(df_land_use_transitions_pantanal.location.unique()) == len(list_municipios_pantanal)
```

```python
dict_rename_transition = {str(i) + "-" + str(i+1): pd.to_datetime(str(i)) for i in range(1985, 2020)}
relevant_cols = ["city", "state", "location","from_level_0", "from_level_1", "from_level_2", "to_level_0", "to_level_1", "to_level_2", "to_level_3", "to_level_4"]
df_land_use_transitions_pantanal = df_land_use_transitions_pantanal[relevant_cols + list(dict_rename_transition.keys())]
df_land_use_transitions_pantanal
```

```python
(
df_land_use_transitions_pantanal[relevant_cols + ["2017-2018", "2018-2019", "2019-2020"]]
.query("from_level_0 == 'Natural' and to_level_0 == 'Anthropic'")
.query("city=='Corumbá'")
).sort_values("2017-2018")
```

```python
melted_df = (
        pd.melt(
            df_land_use_transitions_pantanal.query("from_level_0 == 'Natural' and to_level_0 == 'Anthropic'"),
            id_vars=relevant_cols,
            value_vars=list(dict_rename_transition.keys()),
            var_name="year",
            value_name="deforestation_ha",
        )
        .groupby(["city", "state", "location", "year"])
        .sum()
        .reset_index()
        .assign(year= lambda x: x["year"].replace(dict_rename_transition))
        .rename(columns={"state": "location_UF"})
    )

melted_df
```

```python
io.save_table(melted_df, "domain", "deforestation")
```
